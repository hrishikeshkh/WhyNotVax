```python
import pandas as pd
import matplotlib as plt
df = pd.read_csv('transformed_train_val.csv')
```

```python
class_counts = df['multi_class_labels'].value_counts()

# Print the counts
print(class_counts)

# Plotting the distribution
class_counts.plot(kind='bar')
plt.title('Number of Rows in Each Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
